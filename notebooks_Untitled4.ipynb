Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
     of 5 using Python. Interpret the results.

```python
import scipy.stats as stats

sample_mean = 50
sample_stddev = 5


sample_size = 100


confidence_level = 0.95


standard_error = sample_stddev / (sample_size ** 0.5)


z_value = stats.norm.ppf((1 + confidence_level) / 2)  
margin_of_error = z_value * standard_error


lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error


print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")


"""
      scipy.stats.norm.ppf is used to find the critical z-value for 
      a two-tailed test at the 95% confidence level.
      
    The standard error is calculated as sample_stddev / sqrt(sample_size) 
      since we're assuming a sufficiently large sample size.
      
     The margin of error is calculated by multiplying the standard error
     by the critical z-value.
     
     Finally, the confidence interval is calculated by adding and subtracting
     the margin of error from the sample mean.


   You can be 95% confident that the true population mean falls within the interval provided by the confidence interval.
    In this specific case, it means you can be 95% confident that the true population mean is between approximately
    48.62 and 51.38 based on the sample data. This range represents the uncertainty associated with estimating the
    population mean from a sample, and it's a common way to communicate the precision of your estimate."""
```

Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.

```python
import numpy as np
import scipy.stats as stats


observed_counts = np.array([18, 22, 19, 9, 12, 20])


expected_proportions = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20])


total_observations = np.sum(observed_counts)


expected_counts = expected_proportions * total_observations


chi_square_statistic, p_value = stats.chisquare(f_obs=observed_counts, f_exp=expected_counts, ddof=0)


alpha = 0.05


if p_value < alpha:
    print("Reject the null hypothesis: The distribution of M&M colors does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The distribution of M&M colors matches the expected distribution.")


print(f"Chi-square Statistic: {chi_square_statistic}")
print(f"P-value: {p_value}")
"""
observed_counts represents the actual counts of each color you observed in the M&M bag.
expected_proportions represents the expected proportions of each color based on the provided distribution.
total_observations is the total number of M&Ms in the sample.
expected_counts are the expected counts calculated from the proportions.
The stats.chisquare function is used to perform the chi-square goodness of fit test, 
comparing the observed counts to the expected counts (with ddof=0 indicating no degrees of freedom adjustment)

The code then checks whether to reject the null hypothesis based on the p-value and the chosen significance 
level (0.05). If the p-value is less than the significance level, you reject the null hypothesis,
indicating that the observed distribution differs significantly from the expected distribution."""
```

Q3. Use Python to calculate the chi-square statistic and p-value for a contingency table with the following
data:  Outcome 1 20 15
Outcome 2 10 25
Outcome 3 15 20 
Interpret the results of the test.?

```python
import numpy as np
import scipy.stats as stats


observed_data = np.array([[20, 15],
                          [10, 25],
                          [15, 20]])


chi_square_statistic, p_value, _, _ = stats.chi2_contingency(observed_data)


alpha = 0.05


if p_value < alpha:
    print("Reject the null hypothesis: There is a significant association between the two variables.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between the two variables.")


print(f"Chi-square Statistic: {chi_square_statistic}")
print(f"P-value: {p_value}")
"""
observed_data represents the observed frequencies for each outcome and category in the contingency table.

The stats.chi2_contingency function is used to perform the chi-square test for independence.
It returns the chi-square statistic, p-value, degrees of freedom, and expected frequencies.

The code then checks whether to reject the null hypothesis based
on the p-value and the chosen significance
level (0.05). If the p-value is less than the significance level,
you reject the null hypothesis, indicating
that there is a significant association between the two variables.

If the p-value is greater than the significance level, you fail to reject the null hypothesis, 
suggesting that there is no significant association between the two variables.

Please note that this test assesses whether there is an association between two categorical 
variables but does not imply causation."""
```

Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


df = 10


x = np.linspace(0, 30, 1000)


pdf = stats.chi2.pdf(x, df)


plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label=f'Chi-Square Distribution (df={df})', color='blue')


x_shaded = np.linspace(0, 15, 1000)
plt.fill_between(x_shaded, 0, stats.chi2.pdf(x_shaded, df), where=(x_shaded <= 15), color='gray', alpha=0.5, label='Shaded Area')


plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density Function')
plt.title('Chi-Square Distribution with Shaded Area')
plt.legend()
plt.grid(True)


plt.show()
"""
df represents the degrees of freedom (in this case, 10 degrees of freedom).
We generate a range of chi-square values using np.linspace.
The probability density function (PDF) of the chi-square distribution is calculated using stats.chi2.pdf.
We create a plot using matplotlib and label the axes.
The fill_between function is used to shade the area corresponding to a chi-square statistic of 15."""
```
